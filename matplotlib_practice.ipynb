```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# create the figures
fig, ax = plt.subplots()

# define the axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
fig = plt.figure()
```

```python
plt.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
plt.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.xlim(0,5)
plt.ylim(0,45)
plt.title("Pyber Fare by Month")
plt.grid()
plt.legend("Boston")
```

```python
plt.bar(x_axis, y_axis, color="green",label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color="green",label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color="magenta",label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="orange", label="New York")
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
plt.plot(x_axis, y_axis,"o")
```

```python
plt.scatter(y_axis, x_axis, color="red",label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.xlim(5)
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis, alpha=0.2, linewidth= 2, facecolor = "skyblue", edgecolors="black")
plt.gca().invert_yaxis()
plt.xlim(5)
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
```

```python

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0.3, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
